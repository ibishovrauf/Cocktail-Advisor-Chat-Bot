```python
# check if the len of ingredients and ingredientMeasures are the same in the whole dataset
for _, row in df.iterrows():
    if not (len(eval(row["ingredients"])) == len(eval(row["ingredientMeasures"]))):
        print("cart")
```

```python
for _, row in df.iterrows():
    doc = {
        'name': row['name'],
        'category': row['category'],
        'glassType': row['glassType'],
        'instructions': row['instructions'],
        'ingredients': row['ingredients'],
        'ingredientMeasures': row['ingredientMeasures'],
        'alcoholic': row['alcoholic'],
        'text': row['text'],
        'content': f"   Cocktail: {row['name']}\n \
                        Category: {row['category']}\n \
                        Glass Type: {row['glassType']}\n \
                        Instructions: {row['instructions']}\n \
                        Ingredients: {row['ingredients']}\n \
                        Ingredient Measures: {list(zip(eval(row['ingredients']), eval(row['ingredientMeasures'])))}\n \
                        Alcoholic: {row['alcoholic']}"
    }
    print(doc)
    break
```
